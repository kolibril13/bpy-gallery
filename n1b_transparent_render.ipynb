```python
import bpy
from IPython.display import display, Image

# Set render resolution
bpy.context.scene.render.resolution_x = 500
bpy.context.scene.render.resolution_y = 200

# Enable RGBA (including alpha channel) and set the alpha mode to 'TRANSPARENT'
bpy.context.scene.render.image_settings.file_format = 'PNG'
bpy.context.scene.render.image_settings.color_mode = 'RGBA'
bpy.context.scene.render.film_transparent = True


bpy.ops.render.render()
bpy.data.images['Render Result'].save_render(filepath="img.png")
display(Image(filename="img.png"))
```

```python
import bpy
from IPython.display import Image, display

# Ensure the Cube object is selected and active
obj = bpy.data.objects.get("Cube")
bpy.context.view_layer.objects.active = obj
obj.select_set(True)

# Create and assign material
mat = bpy.data.materials.new("Blue")
mat.use_nodes = True
mat.node_tree.nodes['Principled BSDF'].inputs['Base Color'].default_value = (0, 0, 1, 1)
obj.data.materials.clear()  # Clear any existing materials
obj.data.materials.append(mat)

# Set render resolution
bpy.context.scene.render.resolution_x = 500
bpy.context.scene.render.resolution_y = 200

# Render and display the image
bpy.context.scene.render.filepath = "img.png"
bpy.ops.render.render(write_still=True)
display(Image(filename="img.png"))
```
