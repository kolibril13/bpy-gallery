```python
import bpy
from IPython.display import display, Image

def fresh_scene():
    # Deselect all objects
    bpy.ops.object.select_all(action='DESELECT')
    # Select all objects except lights and cameras
    for obj in bpy.context.scene.objects:
        if obj.type != 'LIGHT' and obj.type != 'CAMERA':
            obj.select_set(True)
        else:
            obj.select_set(False)

def set_resolution():
    bpy.context.scene.render.resolution_x = 500
    bpy.context.scene.render.resolution_y = 200

def render_result():
    bpy.ops.render.render()
    bpy.data.images['Render Result'].save_render(filepath="img.png")
    display(Image(filename="img.png"))

set_resolution()
```

```python
fresh_scene()

# Add a cube
bpy.ops.mesh.primitive_cube_add(size=2, location=(0, 0, 0))

# Add a UV sphere
bpy.ops.mesh.primitive_uv_sphere_add(radius=1, location=(3, 0, 0))

# Add a cone
bpy.ops.mesh.primitive_cone_add(radius1=1, depth=2, location=(6, 0, 0))

# Add a cylinder
bpy.ops.mesh.primitive_cylinder_add(radius=1, depth=2, location=(9, 0, 0))

# Add a torus
bpy.ops.mesh.primitive_torus_add(major_radius=1, minor_radius=0.3, location=(12, 0, 0))


render_result()
```
